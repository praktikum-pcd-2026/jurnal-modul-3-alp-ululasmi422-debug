```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('backup.jpg', cv2.IMREAD_GRAYSCALE)

smooth = cv2.blur(img, (5,5))

kernel = np.array([
    [ 0, -1,  0],
    [-1,  5, -1],
    [ 0, -1,  0]
])

sharp = cv2.filter2D(img, -1, kernel)

hasil = cv2.filter2D(smooth, -1, kernel)

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(smooth, cmap='gray')
plt.title('Smoothing')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(sharp, cmap='gray')
plt.title('Sharpening')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(hasil, cmap='gray')
plt.title('Smoothing + Sharpening')
plt.axis('off')

plt.show()
```

smoothing digunakan untuk menghaluskan gambar dan mengurangi noise.
sharpening digunakan untuk mempertajam detail dan tepi gambar.
sedangkan smoothing + sharpening menggabungkan keduanya agar gambar lebih bersih tetapi tetap tajam.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('backup.jpg', cv2.IMREAD_GRAYSCALE)

prewitt_x = np.array([
    [-1,0,1],
    [-1,0,1],
    [-1,0,1]
])

prewitt_y = np.array([
    [-1,-1,-1],
    [0,0,0],
    [1,1,1]
])

sobel_x = np.array([
    [-1,0,1],
    [-2,0,2],
    [-1,0,1]
])

sobel_y = np.array([
    [-1,-2,-1],
    [0,0,0],
    [1,2,1]
])

roberts_x = np.array([
    [1,0],
    [0,-1]
])

roberts_y = np.array([
    [0,1],
    [-1,0]
])

prewitt = cv2.filter2D(img, -1, prewitt_x) + cv2.filter2D(img, -1, prewitt_y)

sobel = cv2.filter2D(img, -1, sobel_x) + cv2.filter2D(img, -1, sobel_y)

roberts = cv2.filter2D(img, -1, roberts_x) + cv2.filter2D(img, -1, roberts_y)

prewitt_sobel = cv2.add(prewitt, sobel)

prewitt_roberts = cv2.add(prewitt, roberts)

sobel_roberts = cv2.add(sobel, roberts)

gabungan = cv2.add(prewitt_sobel, roberts)

plt.figure(figsize=(14,7))

plt.subplot(2,4,1)
plt.imshow(prewitt, cmap='gray')
plt.title('Prewitt')
plt.axis('off')

plt.subplot(2,4,2)
plt.imshow(sobel, cmap='gray')
plt.title('Sobel')
plt.axis('off')

plt.subplot(2,4,3)
plt.imshow(roberts, cmap='gray')
plt.title('Roberts')
plt.axis('off')

plt.subplot(2,4,4)
plt.imshow(prewitt_sobel, cmap='gray')
plt.title('Prewitt + Sobel')
plt.axis('off')

plt.subplot(2,4,5)
plt.imshow(prewitt_roberts, cmap='gray')
plt.title('Prewitt + Roberts')
plt.axis('off')

plt.subplot(2,4,6)
plt.imshow(sobel_roberts, cmap='gray')
plt.title('Sobel + Roberts')
plt.axis('off')

plt.subplot(2,4,7)
plt.imshow(gabungan, cmap='gray')
plt.title('Prewitt + Sobel + Roberts')
plt.axis('off')

plt.show()
```

operator Prewitt, Sobel, dan Roberts digunakan untuk mendeteksi tepi pada gambar dengan karakteristik yang berbeda. Prewitt bagus untuk mendeteksi tepi sederhana, Sobel lebih baik dalam memperjelas tepi karena memperhitungkan gradien, sedangkan Roberts lebih sensitif terhadap perubahan intensitas yang tajam dan detail kecil.
Kombinasi operator dilakukan agar hasil deteksi tepi menjadi lebih lengkap dan jelas. Penggabungan beberapa operator dapat menonjolkan berbagai jenis tepi yang mungkin tidak terlihat jika hanya menggunakan satu operator saja, sehingga hasil citra menjadi lebih detail dan jelas.


```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('backup.jpg', cv2.IMREAD_GRAYSCALE)

edge = cv2.Canny(img, 100, 200)

_, th10 = cv2.threshold(edge, 10, 255, cv2.THRESH_BINARY)
_, th20 = cv2.threshold(edge, 20, 255, cv2.THRESH_BINARY)
_, th30 = cv2.threshold(edge, 30, 255, cv2.THRESH_BINARY)
_, th40 = cv2.threshold(edge, 40, 255, cv2.THRESH_BINARY)
_, th50 = cv2.threshold(edge, 50, 255, cv2.THRESH_BINARY)
_, th60 = cv2.threshold(edge, 60, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(12,7))

plt.subplot(2,3,1)
plt.imshow(th10, cmap='gray')
plt.title('Threshold 10')
plt.axis('off')

plt.subplot(2,3,2)
plt.imshow(th20, cmap='gray')
plt.title('Threshold 20')
plt.axis('off')

plt.subplot(2,3,3)
plt.imshow(th30, cmap='gray')
plt.title('Threshold 30')
plt.axis('off')

plt.subplot(2,3,4)
plt.imshow(th40, cmap='gray')
plt.title('Threshold 40')
plt.axis('off')

plt.subplot(2,3,5)
plt.imshow(th50, cmap='gray')
plt.title('Threshold 50')
plt.axis('off')

plt.subplot(2,3,6)
plt.imshow(th60, cmap='gray')
plt.title('Threshold 60')
plt.axis('off')

plt.show()
```

 threshold digunakan untuk menentukan batas piksel yang dianggap sebagai tepi pada gambar. Threshold rendah menghasilkan lebih banyak tepi sehingga detail gambar lebih terlihat, tetapi noise juga ikut muncul. Sebaliknya, threshold tinggi membuat hasil lebih bersih karena noise berkurang, namun beberapa detail tepi dapat hilang.
Pada kasus ini, threshold sedang seperti 20 atau 30 biasanya memberikan hasil terbaik karena detail tepi masih terlihat jelas tanpa terlalu banyak noise.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('backup.jpg')
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

_, mask = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY_INV)

kernel = np.ones((5,5), np.uint8)
mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)

edge = cv2.Canny(mask, 100, 200)

hasil = rgb.copy()

hasil[mask == 255] = [255,255,0]

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(mask, cmap='gray')
plt.title('Mask')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(edge, cmap='gray')
plt.title('Edge')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(hasil)
plt.title('Highlight Siluet')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('david.jpg')
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

r, g, b = cv2.split(rgb)
lower_toska = np.array([0, 120, 120])
upper_toska = np.array([70, 255, 255])
mask_bg = cv2.inRange(rgb, lower_toska, upper_toska)

mask_bg = cv2.medianBlur(mask_bg, 7)
kernel = np.ones((5,5), np.uint8)
mask_bg = cv2.morphologyEx(mask_bg, cv2.MORPH_CLOSE, kernel)
mask_bg = cv2.morphologyEx(mask_bg, cv2.MORPH_OPEN, kernel)
mask_fg = cv2.bitwise_not(mask_bg)

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_fg = cv2.bitwise_and(gray, gray, mask=mask_fg)
blur = cv2.GaussianBlur(gray_fg, (3,3), 0)
edge = cv2.Canny(blur, 25, 90)
edge[edge > 0] = 255

hasil_akhir = np.full_like(rgb, 255)
hasil_akhir[mask_fg == 255] = rgb[mask_fg == 255]

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(edge, cmap='gray', vmin=0, vmax=255)
plt.title('Deteksi Tepi')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(mask_fg, cmap='gray', vmin=0, vmax=255)
plt.title('Mask Background')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(hasil_akhir)
plt.title('Hasil Akhir')
plt.axis('off')

plt.show()
```
